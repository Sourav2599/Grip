# Task 6: Prediction Using Decision Tree Algorithm (GRIP)

```python
from sklearn import tree
import pandas
import graphviz
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
df = pandas.read_csv("Iris.csv")
print(df)
```

```python
d={'Iris-setosa': 0,'Iris-versicolor': 1,'Iris-virginica': 2}
df['Species'] = df['Species'].map(d)
print(df)
```

#### 0,1,2 are assigned for class Iris-setosa,Iris-versicolor and Irsi-virginica respectively

```python
features=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
X=df[features]
Y=df['Species']
print(X)
print(Y)
```

## Text Representation 

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, Y)
text_representation=tree.export_text(dtree)
print(text_representation)
```

## Binary Tree Representation

```python
fig=plt.figure(figsize=(25,20))
graph=tree.plot_tree(dtree,feature_names=features,filled=True)
```

```python
print(dtree.predict([[5,3,2,1]]))
```
